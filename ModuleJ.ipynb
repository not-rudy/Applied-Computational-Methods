**Networks Library:** Python has a *Networks* Library used to enable easy plotting of graphs.

```python
## An Example using Networks Library

import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)


# Add edges
G.add_edge(1, 4)
G.add_edge(1, 3)
G.add_edge(3, 4)

G.add_edge(2, 3)
G.add_edge(1, 2)
G.add_edge(4, 2)

# Plot
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.title("Directed Graph")
plt.show()
```

**Laplacian Matrix (L):**

L = D − M, where M is the adjacency matrix and D is the diagonal degree matrix with Dii = ∑k Eik

L is symmetric, positive, semi-definite. All eigenvalues are real and non-negative, and L has a full set of n real and orthogonal eigenvectors

If the graph G has c connected components, then L has c eigenvalues that is 0
For any vector x, x^T Lx = ∑{i, j}∈E Eij(xi − xj)^2

For any vector x and scalars α, β , (αx +βe)^T * L(αx +β e) = α^2(x^T)Lx
