**22.1: ODE Initial Value Problem Statement**

**Differential Equation:** A relationship between a function f(x), its independent variable, x, and any number of its derivatives.

**Ordinary Differential Equation (ODE):** A differential equation where the independent variable and its derivatives are in one dimension.

where F is an arbitrary function that incorporates one or all of the input arguments, and **n** is the order of the differential equation. This equation is referred to as an **nth order ODE**.

**Partial Differential Equation (PDE):** General form differential equation where x is a vector containing the independent variables x1, x2, x2, ... , xm, and the partial derivatives can be of any order with respect to any combination of variables.

**General Solution:** a g(x) that satisfies the differential equation.

**Particular Solution:** a p(x) that satisfies the differential equation and has n explicitly known values of the solution, or its derivatives at certain points.

**P(x) must satisfy the differential equation**

**Initital Value Problem:** Finding a solution to an ODE given an initial value.

**Initial Value:** A common set of known values for an ODE solution

**22.2: Reduction of Order**

**Reduction of Order:** To reduce the order of a differion equation, consider a vector, S(t), which is the state of the system as a function of time. In general, the state of a system is a collection of all the dependent variables that are relevant to the behavior of the system.

**To Reduce:** Turn an nth order ODE into a 1st order ODE

**Linear ODE:** Written in the matrix form of; dS(t)/dt = [ 0, 1; -g/l, 0]S(t)

State of the system can hold multiple dependent variables and their derivaties as long as the derivatives are with respect to the same independent variable.



**22.3: The Euler Method**

**Explicit Euler Formula:** Let dS(t)/dt = F(t, S(t)) be an explicitly defined first order ODE, that is, F is a function that returns the derivative, or change, of a state given a time and state value. Also, let t be a numerical grid of the interval [t0, tf] with spacing h.

Without loss of generality, we can assume that t0 = 0, and that tf = Nh for some positive integer, N.

```python
## An example using the Euler Method

import numpy as np

## Defining Parameters

f = lambda t, s: np.exp(-t) #This is our ODE

h = 0.1 # This is our step size

t = np.arange(0, 1+h, h) # This is our Numerical Grid

s0 = -1 # Initial Condition

## Explicit Euler Method

s = np.zeros(len(t))
s[0] = s0

for i in range(len(t)-1):
    s[i+1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, "b-", label = "Approximate")
plt.plot(t, -np.exp(-t), "g", label = "Exact")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Approximate and Exact Solution for Simple ODE")
plt.legend(loc = "lower right")
plt.grid()
plt.show()
```

```python
## Decreasing Step Size  for Accuracy

import numpy as np

## Defining Parameters

f = lambda t, s: np.exp(-t) #This is our ODE

h = 0.01 # This is our step size

t = np.arange(0, 1+h, h) # This is our Numerical Grid

s0 = -1 # Initial Condition

## Explicit Euler Method

s = np.zeros(len(t))
s[0] = s0

for i in range(len(t)-1):
    s[i+1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, "b-", label = "Approximate")
plt.plot(t, -np.exp(-t), "g", label = "Exact")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title("Approximate and Exact Solution for Simple ODE")
plt.legend(loc = "lower right")
plt.grid()
plt.show()
```
