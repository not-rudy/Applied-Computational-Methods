**21.1 Numerical Integration Problem Statement**

**Numerical Integration Problem Statement:** When given a function f(x), approximate the integral of f(x) over the total **interval**, [a, b]

This is accomplished by assuming that the interval has been discretized into a numeral grid, x, consisting of n+1 points with spacing h. where h = (b-a)/n

Here, we dewnote each point in x by xi, where x0 = a, and xn = b.

There are n+1 grid points due to the starting point being x0.

If we assume that the function f(x) can be computed for any of the grid points, the interval [xi, xi+1] is our **subinterval**


**21.2 Riemann Integral**

**Riemann Integral:** The simplest method for approximating integrals.

This is done by summing the area of rectangles that are defined for each subinterval. The width of the rectangle is xi+1 -xi = h, and the height is defined by a function value f(x) for some x in the subinterval.

This is visualized with image.png

This is based on whether left or right endpoints are chosen for the integration approximation.

**Midpoint Rule:** Takes the rectangle height of the rectangle at each subinterval to be the function value at the midpoint between xi and xi+1. We can comapct the midpoint rule function value to yi = (xi+1 +xi)/2

This follows:

image.png

```python
## An Example Incorporating Riemann Integral and Midpoint Rule

import numpy as np ## Necessary Libraries

a = 0
b = np.pi
n = 11
h = (b - a) / (n-1)
x = np.linspace(a, b, n)
f = np.sin(x)

int_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - int_riemannL

int_riemannR = h * sum(f[1::])
err_riemannR = 2 - int_riemannR

int_mid = h * sum(np.sin((x[:n-1] + x[1:])/2))
err_mid = 2 - int_mid

print(int_riemannL)
print(err_riemannL)
print(int_riemannR)
print(err_riemannR)
print(int_mid)
print(err_mid)
```

**21.3 Trapezoidal Rule**

**Trapezoid Rule:** This rule fits a trapezoid into each subinterval and sums areas of the trapezoids to approxinate the total integral. For each subinterval, the trapezoid rule computes the area of a trapezoid with corners at (xi, 0), (xi+1, 0), (xi, f(xi)), and (xi+1, f(xi+1)), which is h(f(xi)+f(xi+1))/2

Therefore we are left with the expression:

```python
## An Example using Trapezoidal Rule

import numpy as np

a = 0
b = np.pi
n = 11
h = (b-a)/ (n-1)
x = np.linspace(a,b,n)
f = np.sin(x)

I_trap = (h/2) * (f[0] + 2 * sum(f[1:n-1]) + f[n-1])
err_trap = 2 -I_trap

print(I_trap)
print(err_trap)
```
